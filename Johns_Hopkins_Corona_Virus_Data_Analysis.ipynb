```
import re
```

```
english_check = re.compile(r'[a-z]')

def get_state(lon,lat):
  try:
    result = gmaps.reverse_geocode((lat,lon))[0]['formatted_address'].split(",")[-2]
    if(len(result[0].split(" ")) > 1):
      result = "WRST"

    if not english_check.match(result.lower()):
      result = "WRST"

  except:
    print("Error in Reverse Geocoding")
    result = "WRST"
  return result
```
